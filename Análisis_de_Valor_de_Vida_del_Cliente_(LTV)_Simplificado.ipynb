Una app me ha contratado para saber quienes son sus mejores cliente desde su base de datos y por qué.

```python
import pandas as pd
import numpy as np

# Creamos datos ficticios de una App de suscripción
np.random.seed(42)
n_clientes = 200
data = {
    'cliente_id': range(1, n_clientes + 1),
    'meses_antiguedad': np.random.randint(1, 24, n_clientes),
    'gasto_mensual': np.random.uniform(10, 100, n_clientes),
    'soporte_tecnico_clicks': np.random.poisson(2, n_clientes),
    'dispositivo': np.random.choice(['iOS', 'Android', 'Web'], n_clientes)
}
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

Para saber quienes son los clientes VIP, calculamos el gasto mensual por los meses de antiguedad por usuario y ese resultado tiene que ser mayor a su promedio.

```python
df['LTV_estimado'] = df['gasto_mensual'] * df['meses_antiguedad']
df['Perfil_Cliente'] = np.where(df['LTV_estimado']>df['LTV_estimado'].mean(),'VIP','Estandar')
```

```python
df['Perfil_Cliente'].value_counts()
```

```python
df['dispositivo'].value_counts()
```

```python
# ¿Qué dispositivo prefieren nuestros VIPs?
df.groupby('dispositivo')['LTV_estimado'].mean().sort_values(ascending=False)
```

Hemos identificado 84 clientes VIP. Esto ayuda a la empresa saber quienes son y poner atencion a estos usuarios para que sigan siendo clientes VIP ofreciendoles promociones de descuento por clientes leales.
