# Control de Calidad en la Línea de Producción

Tu misión (30 min):
Filtro Crítico: Crea un nuevo DataFrame con las piezas que tengan un error mayor a 0.1 mm (esas son las piezas defectuosas).

Análisis por Operador: Calcula el error promedio por Operador. ¿Quién es el más preciso y quién necesita más entrenamiento?

Visualización Rápida: Haz una gráfica de líneas (.plot()) que muestre cómo varía el error a lo largo de las 10 piezas.

Bonus (Ingeniería): Si el error promedio general es mayor a 0.05 mm, imprime un mensaje de: "ALERTA: Calibrar actuadores de la máquina".

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de error de precisión (en mm)
data = {
    'Pieza_ID': range(1, 11),
    'Error_mm': [0.01, 0.02, 0.05, 0.12, 0.03, 0.25, 0.02, 0.01, 0.15, 0.04],
    'Operador': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A']
}

df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.head()
```

```python
defective_prod = df[df['Error_mm'] > 0.1]
print(defective_prod)
df.groupby("Operador")['Error_mm'].mean().plot(kind='bar',title="Error promedio por Operador.")
plt.xlabel("Operador")
plt.ylabel("Error")
plt.show()
```

```python
def check_calibrator(df):
  if df['Error_mm'].mean() > 0.05:
    print("ALERTA: Calibrar actuadores")
```

```python
check_calibrator(df)
```
