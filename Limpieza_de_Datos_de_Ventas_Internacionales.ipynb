# 🛠️ Reto: "Limpieza de Datos de Ventas Internacionales"
Imagina que recibes un reporte de ventas de una empresa con sede en Malta que vende productos en toda Europa. El archivo es un desastre: tiene símbolos de moneda, fechas mal formateadas, nombres con espacios extra y valores nulos.

```python
import pandas as pd
import numpy as np

data = {
    'Fecha_Venta': ['2023-01-01', '02/01/2023', '2023-01-03', '2023-01-04', 'nan', '2023-01-06'],
    'ID_Producto': ['  PROD_01 ', 'PROD_02', 'prod_01', 'PROD_03', 'PROD_01', ' PROD_02 '],
    'Precio_Unitario': ['$150.00', '120,50€', '$150.00', 'nan', '150.00', '120.50'],
    'Cantidad': [1, 2, np.nan, 1, 5, 2],
    'Email_Cliente': ['CARLOS@mail.com', 'ana@MAIL.com', '  carlos@mail.com', 'pedro@gmail.com', 'info@tienda.com', 'ANA@mail.com']
}

df_dirty = pd.DataFrame(data)
```

```python
df_dirty.info()
```

```python
df_dirty.head()
```

```python
# Estandarizacion de texto
df_dirty['ID_Producto'] = df_dirty['ID_Producto'].str.strip().str.lower()
df_dirty['Email_Cliente'] = df_dirty['Email_Cliente'].str.strip().str.lower()
```

```python
df_dirty = df_dirty.drop_duplicates(subset=['ID_Producto','Email_Cliente'])
```

```python
df_dirty
```

```python
# Limpieza de monedas y conversión numérica
df_dirty['Precio_Unitario'] = df_dirty['Precio_Unitario'].str.replace('$','').str.replace('€','').str.replace(',','.')
df_dirty['Precio_Unitario'] = pd.to_numeric(df_dirty['Precio_Unitario'].replace('nan',np.nan))
```

```python
df_dirty['Precio_Unitario']
```

```python
# Manejo de fechas y nulos
df_dirty['Fecha_Venta'] = pd.to_datetime(df_dirty['Fecha_Venta'],errors='coerce')
df_dirty['Precio_Unitario'] = df_dirty['Precio_Unitario'].fillna(df_dirty['Precio_Unitario'].median())
```

```python
df_dirty = df_dirty.dropna()
```

```python
df_dirty
```

```python
def col_gasto_total(data,col):
  data.loc[:, 'Gasto_Total'] = np.where(data[col] > 200,'Premium','Regular')
  return data
```

```python
df_dirty = col_gasto_total(df_dirty, 'Precio_Unitario')
```
